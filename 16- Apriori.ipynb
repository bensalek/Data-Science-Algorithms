```python
# Apriori

# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python

# Data Preprocessing
dataset = pd.read_csv('D:/Downloads/Udemy.Machine.Learning.A.Z..Hands/24 Apriori/P14-Apriori-Python/Apriori_Python/Market_Basket_Optimisation.csv', header = None)
```

```python
# in range you should write the index of last raw+1! index of last row was 7500! number of rows 7501!
# range(0,20) are the number of columns! 
transactions = []
for i in range(0, 7501):
    transactions.append([str(dataset.values[i,j]) for j in range(0, 20)])
```

```python
# Training Apriori on the dataset
from apyori import apriori
rules = apriori(transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2)

# Visualising the results
results = list(rules)
```
